# Importing Data

```python
#To import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

# Basic Data Modification

```python
#To display how many rows and columns are there
df.shape
```

```python
#To display top 10 rows of the data that we imported
df.head(10)
```

```python
#To display column details
df.info()
```

```python
#To delete column
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
#To check null values in all columns
pd.isnull(df).sum()
```

```python
#To drop null values
df.dropna(inplace=True)
```

```python
#To change the datatype of a column
df['Amount'] = df['Amount'].astype('int')
```

```python
#To check the column datatype
df['Amount'].dtypes
```

```python
#To check columns in dataframe
df.columns
```

```python
#To rename column
df.rename(columns= {'Marital_Status' : 'Shaadi'})
```

```python
#To find the description of data in DataFrame (i.e. count, mean, std, etc). Works only on numeric columns
df.describe()
```

```python
#To use describe function on specific numeric columns
df[['Age','Marital_Status','Amount']].describe()
```

# Exploratory Data Analysis

### Analysis By GENDER

```python
ax = sns.countplot(x = 'Gender', data=df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x= 'Gender', y= 'Amount', data = sales_gen, hue='Gender')
```

---
#### <span style="color:Green;">From above graphs we see that most of the buyers are females and even purchasing power of females are greater than men</span> 
---

### Analysis By AGE

```python
ax = sns.countplot(data=df, x= 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Age Group')
```

---
#### <span style="color:Green;">From above graphs we see that most of the buyers are of Age Group between 26-35 years female</span> 
---

### Analysis By STATE

```python
#Total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders',hue='State')
```

```python
#Total Amount/Sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount',hue='State')
```

---
#### <span style="color:Green;">From above graphs we see that most of the Orders & Total Sales/Amount are from Uttar Pradesh, Maharashtra, and Karnataka respectively</span> 
---

### Analysis By MARITAL STATUS

```python
ax = sns.countplot(x = 'Marital_Status', data=df, hue='Marital_Status')
sns.set(rc={'figure.figsize':(3,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount',hue='Gender')
```

---
#### <span style="color:Green;">From above graphs we see that most of the Buyers are married(women) and they have high purchasing power</span> 
---

### Analysis By OCCUPATION

```python
ax = sns.countplot(x = 'Occupation', data=df, hue='Occupation')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
purchasing_power = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = purchasing_power, x = 'Occupation', y = 'Amount',hue='Occupation')
```

---
#### <span style="color:Green;">From above graphs we see that most of the Buyers are working in IT, Healthcare and Aviation sector</span> 
---

### Analysis By Product Category

```python
ax = sns.countplot(x = 'Product_Category', data=df, hue='Product_Category')
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category', y = 'Amount',hue='Product_Category')
```

---
#### <span style="color:Green;">From above graphs we see that most of the Sold products are from Food, Clothing and Electronics category</span> 
---

### Analysis By Product ID

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID', y = 'Orders',hue='Product_ID')
```

---
### Conclusion
#### <span style="color:Green;line-height: 1.2;">Married women between age group 26-35 years from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to purchase products from Food, Clothing and Electronics category</span>
